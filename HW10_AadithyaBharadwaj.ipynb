##Section 1: ODE Initial Value Problem
- Problem Statement Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dydt(t, y):
    return -y
```

```python
t0 = 0
y0 = 1
tf = 5
h = 0.1
```

```python
t_values = np.arange(t0, tf + h, h)
y_values = np.zeros(len(t_values))
y_values[0] = y0
```

##Section 2: Euler Method Implimentation
- Plot approximate and solution

```python
for i in range(1, len(t_values)):
    y_values[i] = y_values[i - 1] + h * dydt(t_values[i - 1], y_values[i - 1])
```

```python
t_exact = np.linspace(t0, tf, 100)
y_exact = np.exp(-t_exact)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t_values, y_values, 'bo--', label='Euler Approximation')
plt.plot(t_exact, y_exact, 'r-', label='Exact Solution y = exp(-t)')
plt.title('Euler Method for Solving ODE: dy/dt = -y')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

##Section 3: Reduction of Order
- Example (Convert the second-order pendulum equation: d/dt = -(g/l))

```python
def pendulum_system(t, S, g=9.81, l=1.0):
    theta, omega = S
    dtheta_dt = omega
    domega_dt = -(g / l) * theta
    return np.array([dtheta_dt, domega_dt])
```

```python
S0 = [0.1, 0]  # small angle theta = 0.1 rad, omega = 0
h = 0.05
T = np.arange(0, 10, h)
S = np.zeros((len(T), 2))
S[0] = S0
```

```python
for i in range(1, len(T)): #Euler methiod
    S[i] = S[i - 1] + h * pendulum_system(T[i - 1], S[i - 1])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(T, S[:, 0], label='theta (angle)')
plt.title('pendulum - red of order & euler')
plt.xlabel('time (s)')
plt.ylabel('theta (rad)')
plt.grid()
plt.legend()
plt.show()
```
