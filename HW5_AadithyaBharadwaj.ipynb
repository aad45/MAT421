##Section 1
- Linear Space
- Orthogonality
- Eigenvalues and Eignenvectors


```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Linear Space
a1 = np.array([1, 2])
a2 = np.array([3, 4])

alpha, beta = 7, 9
linear_combination = alpha * a1 + beta * a2

print("linear Combination v1 v2:", linear_combination)
```

```python
#Orthogonality
u = np.array([1, 0])
v = np.array([0, 1])

is_orthogonal = np.dot(u, v) == 0
print("u and v orthogonal?", is_orthogonal)
```

```python
#Eigenvalues and eigenvectors
A = np.array([[7, -9], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

##Section 2
- Linear Regression

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 7, 6, 5])
```

```python
coefficients = np.polyfit(x, y, 1)  # degree 1 for linear regression
linear_model = np.poly1d(coefficients)
```

```python
x_line = np.linspace(1, 5, 100)
y_line = linear_model(x_line)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='data points')
plt.plot(x_line, y_line, color='red', label='linear regression line')
plt.title('linear regression example')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
