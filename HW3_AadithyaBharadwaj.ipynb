##Section 1
- Function to calculate the divided difference (from slides)
- Function to calulate newtons polynomial (from slides)

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

#Function to calculate the divided difference (from slides)
def divided_difference(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

    return coef

#Function to calulate newtons polynomial (from slides)
def newton_polynomial(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]
    for k in range(1, n + 1):
        p = coef[n - k] + (x - x_data[n - k]) * p
    return p
```

##Section 2
- Data for Newton's Interpolation
- Evaluating newtons polynomial
- plot newtons interpolation

```python
#Data for Newton's Interpolation
x_newton = np.array([-3, 0, 1, 4])
y_newton = np.array([2, -1, 3, 5])
a_newton = divided_difference(x_newton, y_newton)[0, :]
```

```python
#Evaluating newtons polynomial
x_eval = np.linspace(-3, 4, 100)
y_eval_newton = newton_polynomial(a_newton, x_newton, x_eval)
```

```python
#Plot newtons interpolation
plt.figure(figsize=(10, 6))
plt.plot(x_newton, y_newton, 'bo', label="data points")
plt.plot(x_eval, y_eval_newton, 'r-', label="newton's interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Newton's Polynomial Interpolation")
plt.legend()
plt.grid()
plt.show()
```

##Section 3
- Data for Lagrange Interpolation
- Calculate Lagrange polynomial
- plot langrange polynomial

```python
#Data for Lagrange Interpolation
x_lagrange = np.array([-2, 1, 3, 5])
y_lagrange = np.array([4, -2, 3, 6])
```

```python
#Calculate Lagrange polynomial
lagrange_poly = lagrange(x_lagrange, y_lagrange)

x_eval_lagrange = np.linspace(-3, 6, 100)
y_eval_lagrange = lagrange_poly(x_eval_lagrange)
```

```python
#plot langrange polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_lagrange, y_lagrange, 'ro', label="data points")
plt.plot(x_eval_lagrange, y_eval_lagrange, 'b-', label="lagrange interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Lagrange Polynomial Interpolation")
plt.legend()
plt.grid()
plt.show()
```
