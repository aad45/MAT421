###Example 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simpsons_rule(f, a, b, n):
    if n % 2 == 1:
        n += 1  # make sure n is even
    x = np.linspace(a, b, n+1)
    width = (b - a) / n
    return (width / 3) * (f(x[0]) + 2 * np.sum(f(x[2:n-1:2])) + 4 * np.sum(f(x[1:n:2])) + f(x[n]))
```

```python
def f1(x):
    return np.exp(-x**2)

a1, b1, n1 = -2, 2, 50
simpsons_result1 = simpsons_rule(f1, a1, b1, n1)
print(f"simpson’s Rule approx for exp(-x^2): {simpsons_result1}")
```

```python
x_vals1 = np.linspace(a1, b1, 100)
y_vals1 = f1(x_vals1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals1, y_vals1, label='f(x) = exp(-x^2)')
plt.fill_between(x_vals1, y_vals1, alpha=0.3)
plt.title("Simpson’s Rule Approximation: exp(-x^2)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

###Example 2

```python
def f2(x):
    return np.sin(x)
```

```python
a2, b2, n2 = 0, np.pi, 50
simpsons_result2 = simpsons_rule(f2, a2, b2, n2)
print(f"simpson’s Rule approx for sin(x): {simpsons_result2}")
```

```python
x_vals2 = np.linspace(a2, b2, 100)
y_vals2 = f2(x_vals2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals2, y_vals2, label='f(x) = sin(x)')
plt.fill_between(x_vals2, y_vals2, alpha=0.3)
plt.title("Simpson’s Rule Approximation: sin(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
