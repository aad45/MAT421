## Section 1: Interpolation Problem Statement
- calculating interpolation
- known data points example and plot

```python
from scipy.interpolate import interp1d, CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [1, 2, 0, 4]

x_new = np.linspace(0, 3, 100)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='data points', markersize=8, color='red')
plt.title('known dp')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

#Section 2: Linear Interpolation

- Calculating the linear interpolation
- Plotting the linear interpolation

```python
linear_interp = interp1d(x, y)
y_linear = linear_interp(x_new)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='data pts', markersize=8, color='red')
plt.plot(x_new, y_linear, '--', label='Linear Interpolation', linewidth=2)
plt.title('linear interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

#Section 3: Cubic Interpolation

- Calculating the cyubic spline interpolation
- Plotting the cubic spline interpolation

```python
cubic_interp = CubicSpline(x, y)
y_cubic = cubic_interp(x_new)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Data Points', markersize=8, color='red')
plt.plot(x_new, y_cubic, '-', label='cubic spline interpolation', linewidth=2)
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
