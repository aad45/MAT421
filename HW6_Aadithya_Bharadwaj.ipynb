##Section 1:
- Continuity and Differentiation functions
- Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 3*x**2 + 2*x

def df(x):
    return 3*x**2 - 6*x + 2
```

```python
x_vals = np.linspace(-2, 4, 100)
y_vals = f(x_vals)
dy_vals = df(x_vals)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='function f(x)')
plt.plot(x_vals, dy_vals, label='derivative fx', linestyle='dashed')
plt.title('function and Its derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

##Section 2:
- Taylor Theorem

```python
def taylor_sin(x, n=5):
    approximation = sum(((-1)**k * x**(2*k+1)) / np.math.factorial(2*k+1) for k in range(n))
    return approximation
```

```python
x_vals = np.linspace(-2*np.pi, 2*np.pi, 100)
y_actual = np.sin(x_vals)
y_taylor = [taylor_sin(x) for x in x_vals]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_actual, label='sin(x)')
plt.plot(x_vals, y_taylor, label='taylor approximation', linestyle='dashed')
plt.title('taylor series approxof sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

##Section 3
- Optimization with Gradient Descent
- Example

```python
def gradient_descent(learning_rate=0.1, iterations=50):
    x = 5  # Initial guess
    history = [x]
    for _ in range(iterations):
        grad = 2*x - 4  # Derivative of f(x) = x^2 - 4x + 4
        x -= learning_rate * grad
        history.append(x)
    return history
```

```python
iterations = 50
x_vals = gradient_descent(iterations=iterations)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(iterations+1), x_vals, marker='o')
plt.title('gradient descent optimization')
plt.xlabel('iteration')
plt.ylabel('x Value')
plt.grid()
plt.show()
```
