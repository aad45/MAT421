##Section 1
- Numerical Integration Problem Statement
- Example function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(-x**2)
```

```python
x_values = np.linspace(-2, 2, 100)
y_values = f(x_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = exp(-x^2)')
plt.fill_between(x_values, y_values, alpha=0.3)
plt.title('numerical integration prob statement')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

##Section 2
- Riemann's integral approximation
- using left and right riemann sum method

```python
def left_riemann_sum(f, a, b, n):
    x = np.linspace(a, b, n+1)[:-1]  #left endp
    width = (b - a) / n
    return np.sum(f(x) * width)
```

```python
def right_riemann_sum(f, a, b, n):
    x = np.linspace(a, b, n+1)[1:]  # right endp
    width = (b - a) / n
    return np.sum(f(x) * width)
```

```python
a, b, n = -2, 2, 50
riemann_left_result = left_riemann_sum(f, a, b, n)
riemann_right_result = right_riemann_sum(f, a, b, n)
print(f'left riemann sum approx: {riemann_left_result}')
print(f'right riemann sum approx: {riemann_right_result}')
```

##Section 3
- Trapezoid rule
- visualization

```python
def trapezoidal_rule(f, a, b, n):
    x = np.linspace(a, b, n+1)
    width = (b - a) / n
    return (width / 2) * np.sum(f(x[:-1]) + f(x[1:]))
```

```python
trapezoid_result = trapezoidal_rule(f, a, b, n)
print(f'trapezoid rule approx: {trapezoid_result}')
```
