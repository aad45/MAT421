##Section 1: Incidence matrix
- Example

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0)])
```

```python
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, edge_color='gray')
plt.title("undirected graph")
plt.show()
```

```python
inc_matrix = nx.incidence_matrix(G, oriented=False).todense()
print("unoriented incidence matrix")
print(inc_matrix)
```

```python
inc_matrix_oriented = nx.incidence_matrix(G, oriented=True).todense()
print("oriented incidence matrix")
print(inc_matrix_oriented)
```

##Section 2: Laplacian matrix
- Example

```python
plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, edge_color='black')
plt.title("graph for laplacian matrix")
plt.show()
```

```python
lap_matrix = nx.laplacian_matrix(G).todense()
print("laplacian matrix:")
print(lap_matrix)
```

```python
eigenvalues = np.linalg.eigvals(lap_matrix)
print("eigenvalues of laplacian matrix")
print(np.round(eigenvalues, 3))
```

```python
spectral_gap = sorted(eigenvalues)[1]
print("Spectral Gap (Algebraic Connectivity):", round(spectral_gap, 3))
```
