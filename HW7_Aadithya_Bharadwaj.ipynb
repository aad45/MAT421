##Section 1:
- Numerical Differentiation Problem Statement
- Example function

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exampole function
def f(x):
    return np.sin(x)
```

```python
x_values = np.linspace(0, np.pi, 100)
y_values = f(x_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = sin(x)')
plt.title('mumerical differentiation prob statement')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

##Section 2:
- Finite Difference Approximating Derivatives
- forward, backward, and central difference methods (from the internet)

```python
def forward_difference(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h
```

```python
def backward_difference(f, x, h=1e-5):
    return (f(x) - f(x - h)) / h
```

```python
def central_difference(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
x_test = np.linspace(0, np.pi, 100)
forward_diff = forward_difference(f, x_test)
backward_diff = backward_difference(f, x_test)
central_diff = central_difference(f, x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, np.cos(x_test), label='analytical derivative cos(x)', linestyle='dotted')
plt.plot(x_test, forward_diff, label='forward difference', linestyle='dashed')
plt.plot(x_test, backward_diff, label='backward difference', linestyle='dashdot')
plt.plot(x_test, central_diff, label='central difference', linestyle='solid')
plt.title('finite difference approx of derivatives')
plt.xlabel('x')
plt.ylabel("derivative approximation")
plt.legend()
plt.grid()
plt.show()
```

##Section 3:
- Approximating Higher Order Derivatives
- taylor series expansion to estimate second derivative

```python
def second_derivative(f, x, h=1e-5):
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)
```

```python
second_deriv = second_derivative(f, x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, -np.sin(x_test), label='analytical second derivative -sin(x)', linestyle='dotted')
plt.plot(x_test, second_deriv, label='numerical second derivative', linestyle='dashed')
plt.title('approximating higher order derivatives')
plt.xlabel('x')
plt.ylabel("second derivative approx")
plt.legend()
plt.grid()
plt.show()
```
